Análisis de dataframe

```python
print(housing.info())
print(housing.head())
print(housing['Suburb'].value_counts())
print(housing.describe())

```

```python
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Conteo por Suburb
housing['Suburb'].value_counts().plot(kind='bar', figsize=(10, 6), color='skyblue')

# Etiquetas en inglés
plt.title('Number of Properties per Suburb')
plt.xlabel('Suburb')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
